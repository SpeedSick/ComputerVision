```python
import matplotlib.pyplot as plt
#import matplotlib.image as img
import numpy as np
import scipy.misc as img
from skimage import feature

from scipy import ndimage
import random
%matplotlib inline



image = img.imread('./photo_3.jpg', mode='F' )

print(image.size)

image2 = feature.canny(image, sigma =20)
image3 = feature.canny(image, sigma=30)


fig = plt.figure(figsize=(15,10))

plt.subplot(221)
plt.imshow(image,cmap='hot')
plt.subplot(222)
plt.imshow(image2,cmap='hot')

plt.subplot(223)
plt.imshow(image3,cmap='hot')

image4 = ndimage.sobel(image)

plt.subplot(224)
plt.imshow(image4,cmap='gray')

plt.show()
```
